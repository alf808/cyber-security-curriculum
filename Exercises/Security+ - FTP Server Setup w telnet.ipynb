#### Setting up FTP Server

* We are going to associate a **static IP address** with our Ubuntu server so that it's IP address doesnt change

* Install `vsftpd` using the following command: `sudo apt install vsftpd`


![apt vspftpd package installs](img/apt-vsftpd-install.png)

* Verify your `vsftpd` installation was successful by running the command: `sudo systemctl status vsftpd`

![vsftpd install verification](img/apt-vsftpd-verify.png)


* We can also see that the default port **21** has been opened on our Ubuntu host if we run a port scan from our Kali machine:


![vsftpd nmap verification](img/vsftpd-nmap-verify.png)


* Make a backup of the origingal configuration file and look at the configuration file for `vsftpd`:

    * `cd /etc/vsftpd`
    * `mv vsftpd.conf vsftpd.conf.bak`
    * `vim vsftpd.conf`
    
* Open Wireshark on your Kali machine and start logging `ftp` traffic with the following filter: `ftp or ftp-data`
    
* Demonstrate ability to login to user account with user credentials from our Kali virtual machine using `telnet` and the default FTP port **21**. Use the following command: `telnet x.x.x.x 21`

![kali telnet ftp user login](img/kali-telnet-ftp-user-login.png)


* Common raw commands for using FTP:

    * HELP
    * USER
    * PASS
    * GET
    * PWD
    * CWD
    * PUT
    
* It is more effective to use an actual FTP client however, install a command line FTP client in your Kali virtual machine with: `apt install ftp -y`
    
    
* Once installed you can FTP to your Ubuntu machine and list the contents of your home directory with the following command(s): `ftp x.x.x.x`


![kali ftp user login](img/kali-ftp-user-login.png)


* You can list files on the remote server and run other common Linux/Unix commands. To run any command on the local shell prefix your command with an exclamation: `!ls`


![kali ftp ls](img/kali-ftp-ls.png)

    
* Enable anonymous access and a standard directory where anonymous users files will be stored by changing the following settings in your `/etc/vsftpd.conf`:

    * `anonymous_enable=YES`
    * `no_anon_password=YES`
    * `anon_root=/var/ftp/`

    
![ftp anon settings](img/ftp-anon-settings.png)


* Test the anonymous FTP access in both the browser and `ftp` client on Kali
