#### VirtualBox NAT Network Setup

![Wireshark Network Tab](img/wireshark-network-tab.png)

![Wireshark NAT Network Settings](img/wireshark-nat-configuration.png)

* Click OK to save your changes and exit the VirtualBox Preferences dialogue box

* Select the Network tab on **ALL** your virtual machines with the following settings:

    * Kali
    * Ubuntu 18.04
    * Windows 10


* Select the `NAT Network` option in the `Attached to:` drop down menu


* Select the `NatNetwork` option


* Click the Advanced option


* Select the `Allow All` option in the `Promiscuous Mode:` drop down menu

#### Setting up SSH Server

* Make sure SSH server is running

* Use one of the networking lab Ubuntu machines to create a few users and generate SSH keys.

#### Setup Port Forwarding for accessing Kali w/ Port forwarding

* Make sure host machine has SSH key generated
* Add NAT network interface on Kali Adapter 2
* Setup port forwarding on guest 5555 => kali 22
* Log into Kali 
* Show that SSH is not enabled by default
    * systemctl status ssh
* Enable SSH to start at startup 
    * systemctl enable ssh
    * systemctl status ssh (see enabled flag set)
    * systemctl start ssh
* Add user SSH key to `~/.ssh/authorized_keys`
* Test SSH connection from host to client
* Stop Kali
* Start Kali in headless mode 
* Test SSH again and run tools against internal network
