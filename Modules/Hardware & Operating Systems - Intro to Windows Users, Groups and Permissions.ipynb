# Intro to Windows Users, Groups and Permissions

#### 1. Topics/Sub-Topics
- Users and Groups in Windows
    - What does a user account consist of?
        - Username/Password
        - Encrypted and stored by OS 
        - Knows as Local User Account
    - What is the SYSTEM user account?
        - All processes have to be owned by some user account
    - What is the difference between authentication and authorization? (Sec+)
        - Authentication is the process of identifying and granting access to a system. Prove they are who they say.
        - Authorization is the process of giving users access to individual resources
    
    - What are strong passwords?
    - What is a password expiration policy? (Sec+)
        - Min length
        - Password expiration policy
        - Special characters
        - Non-blank
        - Don't write down
    - What is a group?
    - Why do we use groups?
        - Group people for ease of control and authorization of resources
        - Allow users to be part of multiple groups
    - What are the default Windows groups?
        - Administrators
            - God mode
            - Default for first account created in Windows
        - Power Users
            - Most permission as Administrators except they cannot install new devices or access other users' files or folders without explicit permissions set. Can edit registry and system files
        - Users
            - Cannot edit registry or access system files or other user account files without explicit permissions set
            - Can create groups but only manage those that they create
            - Knows as Standard Users
        - Guests
            - Allows access to a guest account login
            - Best practice is to disable this account (Sec+)
        - Only administrators can create and manage users
    - What is the User Accounts panel?
        - Changes often between Windows versions
        - Found in Control Panel->User Accounts, Windows 7
        - Found in System Settings->Accounts Windows 10
            - Sign-in options to change password, set pin, or picture password
            - Try to force users into using Microsoft accounts
                - Allows for "global" settings and persistence between machines
                - Any valid e-mail address actually works as a "MS address"
        
    - What is Local Users and Groups panel?
        - Found in Administrative Tools in Professional and Enterprise versions
    - What is User Account Control?
        - Enabled as flaw with standard Windows usage
        - Since many people run as admin, it's a safety check from malware being able to run undetected with full access
    - What are Parental Controls?
        - Different in different versions of Windows
        - Can turn on Activity Reporting in some versions to log all activity
        - Restrict overall usage, time of day, etc...
        - Game Rating Systems to block certain kinds of games
    - What are "object types" in Windows?
        - User Accounts
        - Groups
        - Computers
        - Each object type can be added to a group and assigned permissions
    - What is the NTFS file system?
        - NT File System
        - Provides authorization
        - Every folder and file contains 2 sets of data
            - All users and groups that have access
            - The level of access each user or group has
        - Level of access is known as NTFS permissions
            - Ownership - All permissions
            - Take Ownership - Admins have this control
            - Change - Can give or take away permissions from other accounts
            - Folder - Define what a user can do to a folder
            - File - Define what a user can do to a single file
        - Standard folder permissions
            - Full Control
            - Modify
            - Read & Execute
            - List Folder Contents
            - Read
            - Write
        - Standard file permissions
            - Full Control
            - Modify
            - Read & Execute
            - Read
            - Write
        - Security Tab Demo
        - If you have Full Control of a folder, stricter file permissions will not be honored
        - Inheritance
            - New files automatically inherit from parent
            - Bad practice to disable, but show how, recommended to use Deny instead
            - Greyed out permissions in Security tab are inherited
        - **NOTE** Windows home editions only have: Make this folder private and don't have file level permissions
        - Copy/Move Permissions aka Permission Propagation
            - 4 possible scenarios
                - Copy within one NTFS volume
                    - Copy inherits permissions from new location
                - Moving within one NTFS volume
                    - Copy retains original permissions
                - Copy between two NTFS volumes
                    - Copy inherits permissions from new location
                - Moving between two NTFS volumes
                    - Inherits permissions from new location
