##### What is netcat?
Netcat is known as the *Swiss Army Knife of networking*. It is a native Unix utility and at it's core a TCP/UDP tool and hence it's place in the networking tools ecosystem. It can create almost any kind of network connection that you would need in a modern network.

##### Why netcat?
Netcat is installed on most Linux distributions by default. Along with it's flexibility, it's a tool that is commonly known to be available.

Netcat is commonly flagged on Windows anit-virus products as malware and quarantined. In those cases telnet can serve as a similar connection testing tool between systems.

##### netcat Basics
Netcat essentially does **two things** very well:

1. Transfer Data
2. Listen for Requests on a Port

** Flags **

- `-l` - Listen on port
- `-v` - Verbose output, good habit to use on most use cases since default feedback is minimal
- `-p` - Port to listen on
- `-e` - Execute command on connection

#### What can netcat do?

- Used in scripts to automate tasks
- Copy a file between two systems with not FTP or SSH:
- Clone a hard drive or partition
- Port scanning
- Pass or recieve a shell to/from another machine

##### Copy Files Between Servers
    - Server: `nc -lp 1234 > some-file.tar.gz`
    - Client: `nc -w 1 xxx.xxx.xxx.xxx 1234 < some-file.tar.gz`
    
##### Clone Hard Drive or Partition
**Standard usage example:**

`dd if=SomePreviousImage.iso of=/dev/sda`

**With netcat:**

1. Start the listener on the new system receiving the image:
    - Clone Server: `nc -l -p 1234 | dd of=/dev/sda`
    
2. Start the `dd` utility on the original server and pipe stdout to the connection on listening connection:
    - Original Server: `dd if=/dev/sda | nc 192.168.0.12 1234`


##### Port Scanning
- Check for a single port: `nc -v 192.168.1.24 80`
    - What kind of connection is this?
- Check remote server for a range of ports: `nc -v -w 1 192.168.1.24 -z 1-1000`
- Check local server for open ports: `nc -v -w 1 localhost -z 1-1000`
