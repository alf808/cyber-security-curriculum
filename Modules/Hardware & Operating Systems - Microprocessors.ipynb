#### 3. Desired Outcomes
- Students can define the main function of a CPU
- Students understand the concept of binary modes and switches
- Students know that instructions are sent to the CPU via the External Data Bus(EDB)
- Students know what machine language is
- Students understand what the clock wire is and that a charge to the clock wire instructs the CPU to process an instruction set, known as a clock cycle
- Students understand how clock speed is calculated
- Students know what a system crystal and it's purpose is
- Students know what Assemly Language is and how it relates to microprocessors
- Students understand compatibility considerations for a CPU
    - Clock speed of system vs design of chip
- Students understand the concept of overclocking and how it works
- Students understand what registers are and the 4 general purpose registers:
    - General Purpose Registers:
        - AX
        - BX
        - CX
        - DX
    - 32-bit Registers:
        - EAX
        - EBX
        - ECX
        - EDX
    - 64-bit Registers:
        - RAX
        - RBX
        - RCX
        - RDX
- Students understand what memory/RAM is at a high level
- Students understand how and why the CPU and RAM work together
- Students understand what an Address Bus is
- Students understand that RAM transfers instructions to the CPU in 8-bit/byte chunks
- Students know that CPU's have a manufacturer make & model number
- Students should understand the purpose of the MCC(memory controller chip)
- Students should understand how a CPU tells the MMC which area of RAM it wants
- Students know how to calculate the maximum amount of RAM a CPU could handle given the number of wires in CPU's address bus
- Students understand how binary numbers are used for the CPU to access an given "row" or RAM
- Students know that a 32-bit system can only support a maximum of 4GB of RAM
- Students know that a 64-bit system can support up to 16EB of RAM
- Students know the notation of x86, x64, x86-64 and what they represent
- Students know the 4 stages a CPU takes in pipelining:
    - Fetch
    - Decode
    - Execute
    - Write
- Students understand what cache is, each level(L1, L2, L3) and their purpose
- Students understand what multithreading is and what it's purpose is
- Students understand what multicore processing is
- Students understand the difference between multicore processing and multithreading
- Students understand what an Integrated Memory Controller is
- Students understand what an Integrated Graphics Processing Unit is
- Students understand what Nx bit technology is and the term used by different vendors for it's implementation
    - Intel: XD Bit
    - AMD: Enhanced Virus Protection
    - ARM: XN(eXecute Never)
    - CompTIA: Disable execute bit
