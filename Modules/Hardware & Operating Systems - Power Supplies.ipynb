# Hardware & Operating Systems - Power Supplies

#### 1. Topics/Sub-Topics

- The power supply takes electricity from the wall socket and transforms it into electricity to run computer
- Electricity
    - flow of negatively charged electrons through matter
    - Similar to water flowing
    - voltage- pressure of electrons(volts)
    - Current- amount of electrons moving past a certain point on a wire (amps)
    - Wattage- amount of amps and volts needed so that a device will function (watts)
    - Resistance- resistance to flow of electrons (ohms)
    - V x A = W
- Electricity comes in 2 flavors: 
    1. Direct Current (DC)- electrons flow in one direction around a continuous circuit
    1. Alternating Current (AC)- electron alternate direction back and forth in a circuit
- Most electronic devices use DC power but all power companies supply AC because i travels long distances more efficiently
- AC comes in somewhere between 110 and 120 
- Power supply connects to the power cord via IEC-320 connector
- Multimeter- Used to test power to outlets and devices 
- Uninterruptible power supply (UPS)- protects your computer (and data!) in the event of a power sag or power outage
- What do you need to consider when getting the right UPS?
    - You need to calculate how much wattage your computer, monitor, drives, etc require in order to get a proper UPS
    - How long do you want/need power supply to run when you lose power?
    - [Wattage Calculator](http://www.coolermaster.com/power-supply-calculator/)
- Motherboards use a 20- or 24-pin P1 power connector
- SATA Power Connectors

#### 3. Desired Outcomes
- Students should understand that problems with power can create system instability crashes and data loss
- Students should know basic principles of electricity 
- Students should know how to recognize power problems and how to solve them
- **Students should know that a power supply is also referred to a PSU and also falls into the category of field replaceable unit**
- **Students should know that an electrical outlet must have a ground wire to be suitable**
- **Students should be familiar with power supply dual voltage**
- Students should know what a multimeter is and how to use it (pg. 277)
- **Students should be familiar with the technology and use of surge suppressors and battery backup systems (UPS)**

- Students can define the main function of a CPU
- Students understand the concept of binary modes and switches
- Students know that instructions are sent to the CPU via the External Data Bus(EDB)
