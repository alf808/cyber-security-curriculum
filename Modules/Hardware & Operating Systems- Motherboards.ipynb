# Hardware & Operating Systems - Motherboards

#### 1. Topics/Sub-Topics

- What is a motherboard?
    - Every piece of hardware directly or indirectly plugs into motherboard
    - Contains wires (traces) that make up the buses of the system
    - Good picture on pg. 232
- How It Works?
    - Form Factor (determines physical size, and general location of components and ports of MB)
        - Industry standardized shapes and layouts that enable motherboards to work with cases and power supplies
        - Applies to case, motherboard and power supply
        - AT Form Factor
        - ATX Form Factor
        - ITX Form Factor
        - Proprietary Form Factors 
    - Chipset (defines type of processor and RAM motherboard requires)
        - One or more discrete integrated circuit chips that supports CPU’s interfacing to all other devices on MB
    - Components (built in components determine core functionality of system)
        - FireWire, built-in sound, hard drive RAID controllers, network cards, fan power connectors, running lights
- What are Form Factors?
- What defines how the air moves around the case?
- Expansion Bus
    - Expansion slots are used for adding expansion cards and thus new functions to PC
    - The slots and accompanying wires and support chips on the first PC and on the latest and greatest PC 
    - Connect through chipset
- Clock crystal and role in running devices at the same speed
- PCI bus architecture and how it changed PC expansion bus
    - Coexists with other expansion buses
    - Self configuring
    - Keep old expansions cards and slowly migrate to PCI
- Four steps to installing Expansion Cards
    1. Knowledge- Know the card works
    1. Physical Installation- Insert properly without damaging motherboard or card
    1. Device Drivers- Provide proper drivers for the specific OS
    1. Verification- Verify card functions properly before walking away
- How do you install device drivers
    - Using installation disc
    - Add hardware wizard in the control panel (Vista Only)
    - Install driver after installing device or it will not see device
- A properly installed expansion card rarely doesn't work
- The “!” signal is the most common error symbol and usually the easiest to fix
- How do you upgrade and install motherboards?
    - Will CPU work with motherboard?
    - How much RAM do you need and are extra sockets available?
    - Get the right case
- Know the basics of Installing motherboard
- Troubleshoot motherboards 
    - Catastrophic- PC wont boot
    - Component- flaky connections between a device
    - Ethereal- Blue screen of death, reboots itself, power supply problems, faulty components, etc.
- Post Card to diagnose
- Flashing the BIOS

#### 3. Desired Outcomes
*Bold= Exam tip from book*
- Students can define what a motherboard is
- Students understand the 3 parts of a motherboard- form factors, components, and chipset
- Students should know that every device in computer connects to external data bus and address bus.
- Students know PCI and what makes it better
- Students should understand that a PCI express is the most popular and doesn't share the bus and has its own direct connection
- **Students should know the different types of PCI (AGP, PCI-X, Mini-PCI, Mini-PCle, PCI Express)**
- **Students should know the 4 steps involved in installing expansion cards and that those steps apply to all types expansion cards (USB, modem, wireless cards, etc.)**
- Students should know how to install drivers for specific devices
- Students understand the importance of correctly installing devices (expansion cards) 
- Students know signs of improperly installed expansion card and how to troubleshoot common issues
- **Students should be able to select and install a motherboard appropriate for a client or customer**
- **Students should KNOW the basics of installing a motherboard**
- Students should know what to look for when troubleshooting motherboards 
- Students should know the 3 basic types of failures- catastrophic, component, and ethereal
- **Distended Capacitors (pg. 266)**
