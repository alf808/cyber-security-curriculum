# OSI Model

In the beginning, computers could only communicate with computers from the same manufacturer (Like IBM). In the late 1970's OSI reference model was created to break through barrier. Describes how data and network information are communicated from an application on one computer through the network to media. 

The OSI model is important because many certification tests use it to determine your understanding of computer networking concepts.

Used to visualize computer networks

Why is this important/helpful?
- Creates a set of computer networking standards
- Divides the network communication process into smaller and simpler components
- Allows various types of network hardware and software to communicate
- Prevents changes in one layer from affecting another layer

Not a physical model rather its a set of guidelines
Software (How applications within the end stations will communicate with each other and with users)

1. Application (Data)
    - Closest to Human interaction 
    - Application Layer provides Interface between users and machines (application programs).
    - file transfers, email, remote access, client/server processes like printing
    - This is where we users actually communicate to the computer
    - File, print, message, database, and application services
    - Protocols of this layer are: HTTP, HTTPS, FTP, TFTP, Telnet, SNMP, DNS, Rlogin, SMTP, POP3, IMAP, and LDAP.
    - Microsoft word does not reside at the application layer but instead interfaces with the application-layer protocols
2. Presentation (Data)
    - Presents data to the application layer and is responsible for data translation and code formatting
    - Translator
    - This layer facilitates the presentation of Data to the upper layer. Mainly, Provide Encoding Scheme & Encryption formation.
    - Data encryption, compression and translation services
    - Protocols of this layer: JPEG, BMP, GIF, TIF, PNG, MP3, MIDI, ASCII & ANSI etc.
3. Session (Data)
    - Responsible for setting up, managing, and then tearing down sessions between presentation-layer entities
    - Basically keeps applications' data separate from other applications' data
    - This layer provides virtual agreement between two end communication devices.
    - Functions of this layer: Establishment, Management & Termination
    - The best example to explain this layer is telephone call in which first you established the connection, then exchange a message and finally terminate the session.
    - Protocols of this layer: SIP, NFS, SQL, ASP, RDBMS
How Data is Transmitted 
4. Transport
    - This layer is responsible for Control of Data flow and, if an error occurs, reconnect the data and re-transmit.
    - reassembles data into a data stream
    - Reassemble data from upper-layer applications and unite it onto the same data stream
    - End-to-end connection
    - Functions of this layer: Handshaking, Acknowledgment and Sequencing
    - Protocols of this layer: TCP, UDP, SPX
    - Connectionless or connection-oriented (See Protocols Module)
5. Network
    - Routing 
    - Manages device addressing, tracks the location of devices on the network and determines the best way to move data
    - Logical addressing which routers use to determine path
    - This layer is used for communication to remote networks.
    - EX: packet is received on a router interface, the destination IP is checked. If packet isn't for that router, it then looks up destination network address in the routing table. chooses an exit interface and packet is sent. If it cant find location, router drops packet
    - THREE types of packets used at Network Layer (Data Packets, Route-Update packets, Network Addresses)
        - Data Packets
            - IP & IPv6
        - Route-update packets
            - Used to update neighboring routers about the networks connected to all routers
            - RIP
         - Network Addresses
             - Protocol specific network addresses
             - Router retains a routing table for individual routing protocols because each routing protocol keeps track of of a network with a different addressing scheme
  (See Routing Module! pg. 47)
    - Functions of this layer: Sorting, Filtering and Distribution
    - Protocols of this layer: Routed Protocol: IP/IPX/Apple talk
    - Routing Protocol: IGP, EGP, BGP, EBGP, IBGP, RIP, IGRP, RIP, OPSF, IS-IS
6. Data Link
    - Framing
    - Provides the physical transmission of the data and handles error notification, network topology and flow control
    - ensures that messages are delivered to the proper device on the LAN using hardware addresses
    - Also translates messages from the network layer into bits for the physical layer to transmit
    - Routers don't care where a particular host is located they only care about where networks are located and how to reach them
    - This layer is responsible for the unique identification of each device that resides on a local network
    - Each time a packet is sent between routers its framed with control information in the data link layer then stripped off at receiving router
7. Physical (Binary) 
    - Concerned with the transmission and reception of the unstructured raw bit stream over a physical medium
    - Sends and receives bits (binary)
    - specific protocols are needed for each type of media to describe the proper bit patterns to be used
    - The physical layer specifies the electrical, mechanical, procedural and functional requirments for activating, maintaining and deactivating a physical link between end systems
    - Specifies the layout of the transmission media (Its topology in other words)


Troubleshooting
When first discovering a networking problem, it can be confusing as to where you should start. The OSI model helps you to start figuring out where the problem may reside and therefore, help you to solve the problem. 

- For example, a clientâ€™s computer is not able to reach a website. The problem is isolated to one device on the network, so it can be assumed that the problem is likely to be on the application layer. However, after testing you find that the problem still exists when you try a different web browser and so the application layer is ruled out. You then assume that the problem resides on the presentation layer and start looking for faulty settings. After some digging around you find that your client has entered some DNS settings incorrectly and so you then conclude the problem is on the presentation layer.

Encapsulation
When a host transmits data across a network to another device, data is encapsulated
Each Layer communicates only with its peer layer on the receiving device
To communicate and exchange information each layers used Protocol data Units (PDU's)- these hold the control information attached to the data at each layer of the model (Usually attached to the header)
1. User info is converted to data for transmission on network
2. Data is converted to segments and a reliable connection is set up
3. Segments are converted to packets or datagrams and a logical address is placed in the header so each packet can be routed through an internetwork
4. Packets or datagrams are converted to frames for transmission on local network. Hardware addresses are used to uniquely identify hosts on a local network segment
5. Frames are converted to bits and a digital encoding and clocking scheme is used.

The following network devices operate at all seven layers of the OSI model:
- Network management stations (NMSs)
- Web and application servers
- Gateways (not default gateways)
- Network hosts
- Several devices operate primarily at the Physical layer of the OSI model. These devices manipulate mainly the physical aspects of a network data stream (such as the voltages, signal direction, and signal strength). The most popular of these are the following:
- Network Interface Cards (NICs)
- Transceivers
- Repeaters
- Hubs
