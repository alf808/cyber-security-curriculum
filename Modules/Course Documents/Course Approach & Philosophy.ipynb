### Cyber Security Professional Course Philosophy
**Inspiration**
[How To Build a Successful Information Security Career](https://danielmiessler.com/blog/build-successful-infosec-career/#education)

The Cyber Security Professional course will focus on hardening skills in the following areas:

- Networking
- Systems Administration
- Development

### Mission Statement

The Cyber Security Professional course is an immersive training environment designed to prepare individuals for an entry into the field of cybersecurity and to train individuals capable of critical thinking in a variety of technical areas and technical competencies in the area of operating systems, networks, data analysis, development and operational capabilities.

### Learning Environment

The Cyber Security Professional track employs an agile, collaborative and peer focused environment. Instruction will be a blended model of lecture, hands on mentorship, peer collaboration, and project based learning.

### Weekly Approach

Every week, as a guideline, rather than absolute, should be comprised of various composition of the following:

**Lecture**

Post-mortem and introductory, fundamental truths
This track will include lecture but in more meaningful ways than simply walking students through how things work and expecting that information will be retained.

Lecture can happen in a few ways:
1. **Introductory** - Explaining new concepts at a high level and putting context on subject matter for clarity and purpose.
1. **Post-Mortem** - Lectures or Q/A sessions may be more productive after students have had time to work through subject matter and formulate questions, opinions or additional information.
1. **Fundamental Truths** - One method of condensing lecture times but still providing value is to provide a short list of things to know or remember about a given topic. Students can be expected to do depth of research about the minutia of a tool/topic but fundamental truths are things learned through experience not always easily discoverable by early entry.

**Research**

Students should be given guidance for the material that will be covered in a week and keywords that they should focus on. They should also be given ample time to conduct basic research to mentally process information on their own and within teams.

This time should be used to understand high level concepts on a more personal level and formulate personal questions.

**Documentation/Writing**

One of the ways students can retain information and challenge their mental models is by writing about a subject. Writing can help diagnose weak points of understanding as well as get students comfortable using terminolgy and connecting concepts while communicating to another entity.

Students should have an ongoing dictionary of security considerations and best practices.

**Self-Assessment/Backlog Grooming**

Students should always be analyzing their own understanding of subject matter and grooming their personal backlog.

Backlog Master Copy: https://trello.com/b/9zwpxbEN/cyber-student-backlog-master-copy

**Coding**

Coding will be an emphasis in the course. It may not be covered immediately but it will be introduced early and students will be encourged and be given opportunity to continue to grow their coding skills over a longer duration of the course. The ability to code effectively is something that needs to be nurtured rather than covered in a short sprint and expected to be retained in any meaningful way. More advanced students can progress into building tools for just about every subject or trying to contribute to other open source tools.

**Labs**

Labs will be frequent to give students a project based, hands on learning environment for most subject matter. Labs primarily should be able to be run on a local machine with virtual machines or environment if possible, but sometimes may require dedicated resources on property or in the cloud.

**Question List**

Students should have an ongoing list of questions from week to week that they are actively seeking to answer on their own or seeking clarity from instructors or peers.

### Planning Considerations

##### 4-Hour Sessions
Evening sessions often accompany a long day and students may feel a little more burnt out or tired. Lectures should be kept short and interactive if possible. If students can have hands on keyboard or working in groups and communicating it can keep energy levels up.

Long lectures are hard to pay attention to and retain information at the end of a long day.

##### 11-Hour Sessions
Longer sessions start early in the morning and accommodate longer lectures and labs well. Longer lectures may get better retainment but the long days are also good for in-depth labs where students can go deep into a subject without having to "call it quits" for the day.


### Handling Varying Skill Levels
Because we will be doing programming, research and writing throughout the duration of the course, no student should ever not be challenged to be progressing outside of the defined lecture and activities.

Advanced students should work on contributing or building their own open source tools, or actively dig deeper into subject matter and establish their advanced comprehension of subject matter whether that be through tooling, writing or in-depth research.

Instructors can help guide on strategy and more advanced concepts as needed, but students will own their progress. This is when instructors become more mentor than teacher/lecturer, but students should still feel tended too and that they are receiving value and guidance in ways they couldn't do alone.
