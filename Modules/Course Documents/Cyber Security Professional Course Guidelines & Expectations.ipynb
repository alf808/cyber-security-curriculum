# Cyber Security Professional Course Guidelines & Expectations

Aloha and welcome to the DevLeague Cyber Security Professional immersive accelerated learning program. 

We are honored to have you as a part of this journey and we look forward to becoming your stewards on a unique learning experience. We hope to challenge you, help you, guide you, and ultimately watch you develop your own skills and interests and pave your own path towards a career in cybersecurity.

### Mission Statement

The Cyber Security Professional course is an immersive training environment designed to prepare individuals for an entry into the field of cybersecurity and to train individuals capable of critical thinking in a variety of technical areas and technical competencies in the area of operating systems, networks, data analysis, development and operational capabilities.

### Learning Environment

The Cyber Security Professional track employs an agile, collaborative and peer focused environment. Instruction will be a blended model of lecture, hands on mentorship, peer collaboration, and project based learning.

### Core Values

- **Commitment**
    - Commitment to yourself, commit to your goals
- **Perserverance**
    - Things will get hard, it will be ok
- **Respect**
    - We all will respect one another as peers and equals
- **Compassion**
    - We will help you and we expect you to help one another
- **Accountability**
    - Every individual is responsible for their own time and their own progress
- **Curiosity**
    - Be curious always, question the questions and question the answers
- **Document**
    - Document your progress, document your thoughtsm, document your knowledge, document your mistakes
- **Build**
    - The professional who can build is a league apart, it's better to be able to build tools than just use tools

### Instructor Expectations
As your instructors, we understand that we are your stewards and we have a responsibility to provide a conducive and productive educational experience. 

- We commit to do our best to communicate, educate and mentor students in ways that are beneficial and productive to each individual to the maximum extent feasible in ensuring the best possible outcome for you
- We commit to communicating clear objectives and desired outcomes for exercises and modules
- We commit to demonstrate patience, understanding, compassion and guidance throughout the learning experience
- We commit to be honest and objective with communicating individual and group progress and areas of improvement and strength
- We commit to being available as a resource during and after the program, or in the event we cannot provide guidance, connecting you with resources as needed and available
- We commit to being flexible in our methods to best meeting the learning styles of individuals and the cohort as a whole

### Student Expectations
As a student, you understand that you are in an immersive and accelerated learning program. You may have prior educational experience that may or may not be a similar format, but acknowledge you are in a dynamic, project based and hands on environment.

- You commit to ensuring your own outcome to the best of your abilities and understand you are ultimately responsible and accountable for your own goals
- You commit to challenging yourself to be the best version of yourself you can be upon completion of the program
- You commit to putting in the required time in and outside of classroom to complete work and study additional material as needed
- You commit to being accountable for your attendance and punctuality
- You commit to communicating with your peers professionally
- You commit to proactively communicate struggles, frustrations, interests and personal learning preferences to instructors
- You commit to seeking help for additional understanding as you feel is needed
- You commit to taking notes, reviewing content, or pursuing additional materials to reinforce concepts in the way that is most conducive to your style of learning 
- You understand that one of the most vital skills for a successful career in IT is a passion for learning and the ability to learn effectively whether through guidance or self-direction
