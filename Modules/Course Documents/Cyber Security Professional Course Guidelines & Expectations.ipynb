# Cyber Security Professional Course Guidelines & Expectations

Aloha and welcome to the DevLeague Cyber Security Professional immersive accelerated learning program. 

We are honored to have you as a part of this journey and we look forward to becoming your stewards on a unique learning experience. We hope to challenge you, help you, guide you, and ultimately watch you develop your own personal skills and interests and pave your own path towards a meaningful career in cybersecurity.

### Mission Statement

The Cyber Security Professional course is an immersive training environment designed to prepare individuals for an entry into the field of cybersecurity, as well as enhance individuals' capacity for critical thinking and analysis in a variety of technical areas and competencies; specifically in the areas of operating systems, networking, data analysis, penetration testing, tool development and operational capabilities.

### Learning Environment

The Cyber Security Professional track employs an agile, collaborative, and peer focused environment where individuals are pushed to reach the best of their abilities. Instruction will be a blended model of lecture, hands on mentorship, peer collaboration, self directed and project based learning.

### Core Values

- **Commitment**
    - Commitment to yourself, commit to your goals
- **Perserverance**
    - Things will get hard, it will be okay
- **Respect**
    - We all will respect one another as peers and equals
- **Compassion**
    - We will help you and we expect you to help one another
- **Accountability**
    - Every individual is responsible for their own time and their own progress
- **Curiosity**
    - Be curious always, question the questions and question the answers
- **Creativity**
    - The professional who can create tools in addition to maintain or use systems/tools, it's better to be able to build tools than just use tools
- **Tracking**
    - Document your progress, document your thoughts, document your knowledge, document your mistakes

### Instructor Expectations
As your instructors, we understand that we are your stewards and we have a responsibility to provide a conducive and productive educational experience. 

- We commit to do our best to communicate, educate and mentor students in ways that are beneficial and productive to each individual to the maximum extent feasible in ensuring the best possible outcome for you
- We commit to communicating clear objectives and desired outcomes for exercises and modules
- We commit to demonstrate patience, understanding, compassion and guidance throughout the learning experience
- We commit to be honest and objective with communicating individual and group progress while assessing and identifying areas of improvement and strength
- We commit to being available as a resource during and after the program, or in the event we cannot provide guidance, connecting you with resources as needed and available
- We commit to being flexible in our methods to best meeting the learning styles of individuals and the cohort as a whole

### Student Expectations
As a student, you understand that you are in an immersive and accelerated learning program. You may have prior educational experience that may or may not be of a similar format, but acknowledge you are in a dynamic, project based and hands on environment.

- You commit to ensuring your own outcome to the best of your abilities and understand you are ultimately responsible and accountable for meeting your own goals
- You commit to challenging yourself to be the best version of yourself you can be upon completion of the program
- You commit to putting in the required time in and outside of classroom to complete work and study additional material as needed
- You commit to being accountable for your attendance and punctuality
- You commit to communicating with your peers in a professional manner
- You commit to pro-actively communicate struggles, frustrations, interests and personal learning preferences to instructors
- You commit to seeking help for additional understanding as you feel is needed
- You commit to taking notes, reviewing content, or pursuing additional materials to reinforce concepts in the way that is most conducive to your style of learning 
- You understand that one of the most vital skills for a successful career in IT is a passion for learning and the ability to learn effectively whether through guidance or self-direction


### Knowing How To Ask Questions
One of the key traits, specifically for a cybersecurity professional, but even for a career in technology in general, is the ability to have a curious mind. To identify the questions one must ask, and be able to effectively search for answers or solutions. This can manifest itself in learning new tools or concepts, assessing the security posture of an organization, creating new tools, or strategically building a successful career path and find stimulating opportunities.

### Being The Best Version Of Yourself
The goal of your continuing education is focused on your own personal career and outcome, and being the best version of you that you can be, not the best version of someone else. Everyone has their own personal struggles, focus on overcoming your own.

### Accountability
As a last note, it is also crucial that you realize that you are ultimately responsible for your outcome. There is no secret sauce to success in any career path short of hard work and determination. We cannot force you to put in the extra time outside of class, nor can we force you to make this newly chosen path a daily priority in your life. However, if you can commit 
to yourself, that is the best chance you have at getting the outcome you desire. 

### Building Your Personal Portfolio
It is important to remember that the ultimate goal of the DevLeague Cyber Security Professional track is that students achieve a meaningful start to a fulfilling career path as opposed to simply knowing technical subject matter. 

Part of the Cyber Security Professional track will be a focus on helping you build your personal brand and portfolio through the following methods:

1. Ongoing Technical Documentation
1. Professional Tool Development and/or Contribution
1. Online Presence & Outreach

### Commitment To Ethical Responsibility
As part of the DevLeague Cyber Security Professional course, you commit to upholding the highest standards in ethical and responsible behavior. Some of the tools and concepts covered throughout the course can be used for malicious or irresponsible behavior in physical and cyberspace. As a student, you commit to act legally and ethically at all times in and out of the classroom environment, and to not engage in any activities or behavior that would put yourself, your peers, DevLeague or any other party in a compromising or unfavorable position.
