##### Why netstat?
Netstat is installed on most Linux and Windows distributions by default. Netstat is good at showing connections open on your machine. 

##### netcat Basics
Netcat essentially does **two things** very well:

1. Transfer Data
2. Listen for Requests on a Port

** Flags **

- `-l` - Listen on port
- `-v` - Verbose output, good habit to use on most use cases since default feedback is minimal
- `-p` - Port to listen on
- `-e` - Execute command on connection
