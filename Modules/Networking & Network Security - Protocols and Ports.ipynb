### Process Application Layer Protocols

 Telnet
 - Chameleon of protocols
 - terminal emulation
 - allows a user on a remote client machine, called the client, to access the resource sof another machine the telnet server
 - It does this by pulling a fast one on the telent server and making the client machine appear as though it were a terminal directy atatched to the local network
 - Called a software shell - a virtual terminal that can interact wth the chosen remote host
 - Users begin a telnet session by running the Telent client software and then logging into the telenet server
 - Offers no security or encryption and is being replaced by SSH
 
FTP
- File Transfer Protocol
- protocol that actually lets you transfer files accross an IP network
- Its not just a protocol but also a program
- when it operates as a protocol, FTP is used by applications
- As a program its employed by users to preform tasks by hand

SFTP 
- Secure File Tansfer when you need/want to transer files over an encrypted connection 
- Uses SSH
- Apart from the secure part it is just like FTP and is used to transfer files
- EX: when you need to send or recieve a large file from someone you can use FTP
- FTP is faster than email

SMTP
- Simple Mail Transfer
- uses a queued method of mail delivery
- Once a message has been sent to a destination, the message is spooled to a device, usually a disk
- The server software at the destination than regularly checks the queue for messages
- when it detects them it proceeds to deliver them


TLS
- Transport Layer Security
- Cryptographic protocols
- Comes in handy for enabling secure online data transfer activities like web browsing


SIP
- Session Initiation Protocol
- Signaling protocol used to construct and deconstruct multimedia communication sessions for many things like voice and video calls

SNMP
- Simple Network Managment Protocol 
- Colelcts and manipulates valuable network information
- gathers data by polling the devices on the network from a management station at fixed or random intervals
- Recieves a baseline
- Can stand as watchdog over network
- Can help simplify the process of setting up a network


SSH
- Secure Shell
- Sets up a secure Telnet session over a standard TCP/IP connection

HTTP
- Hypertext Transfer Protocol
- used to manage communications between web browsers and web servers and opens the right resource when you click a link 

HTTPS
- Secure version that arms you with a whole bunch of security tools for keeping transactions between web browsers and servers secure


DNS
- Domain Name Service
- resolves hostnames, specifically internet names (like www.google.com) to their corresponding IP address
- DNS makes our lives easier
- EX: If you wanted to move your Website to a different Service provider, the IP address woulf change and no one would know where to find it.
- DNS uses a domain name to specify an IP address

DHCP
- Dynamic Host Configuration protocol
- Assigns IP addresses to hosts with information provide by a server
- Routers are the most common form of hardware
- the DHCP Server can provide a host
    - IP address
    - Subnet mask
    - Domain Name
    - Default gateway
    - DNS
- Client sends out a DHCP discover message in order to recieve an IP address as a broadcast message
    - layer 2 is all Fs in hex
    - layer 3 is 255.255.255.255
    - connectionless so it uses udp
- Apipa (Automatic Private IP addressing)
    - Windows provides it
    - clients can automatically self configure an ip address and subnet mask in the range 169.254.0.1
    - Automatically assigns a class b subnet mask 255.255.0.0
    - If your IP address is in this range your DHCP server is not working
    
- Four step proccess
    1. Client broadcast DHCP discover
    1. Server sends a unicast DHCP offer
    1. Client broadcast DHCP request
    1. Server Unicast DHCP ack



### Transmission Control Protocol (TCP)


TCP/IP first started in 1973, in 1978 it was divided into TCP and IP then back to TCP/IP in 1983

TCP/IP Model (condensed version of the OSI Model)
1. Process/Application Layer -> (Application, Presentation, Session)
2. Host-to-host Layer        -> (Transport)
3. Internet Layer            -> (Network)
4. Network Access Layer      -> (Data Link, Physical)

Process/Application Layer
- Defines protocols for node-to-node application communication
- Controls user-interface specifications
- Telnet, FTP, TFTP, SMTP

Host to host layer 
- Parallels the functions if the OSI's Transport layer, defining protocols for setting up the level of transmission service for applications
- Tackles issues of creating reliable end-to-end communication
- Packet Sequencing and maintains data integrity
- TCP /  UDP

Internet Layer
- Protocols related to logical transmission of packets over the entire network
- Takes care of addressing of hosts by giving them IP addresses 
- Routing of Packets
- ICMP, ARP, IP

Network Access Layer
- Monitors the date exchange between host and network
- oversees hardware addressing and defines protocols for the physical transmission of date
- Ethernet, Fast Ethernet, Token Ring 


UDP 
- USer Datagram Protocol
- Doesnt offer all the bells and whistles of TCP but it is fabulous at transporting information that doesnt require reliable delivery
- Like a skinny person on a bench (doesnt take up a lot of room or bandwidth on a network)
- We use UDP when when reliability is already handled at the process application layer
- Does not sequence the segments and does not care what order the segments arrive in at the destination

- Many DoS attacks use UDP
    - Send large number of UDP packets to random ports on a remote host
    - The victimized system will be forced into sending many ICMP packets, eventually leading it to be unreachable by other clients. 
    - The attacker(s) may also spoof the IP address of the UDP packets, ensuring that the excessive ICMP return packets do not reach them, and anonymizing their network location(s)

Flow Control
- Data integrity is ensured at the transport layer by maintaining flow control
- Provides a means for the receiver to govern the amount of data sent by the sender
- Prevents a sending host on one side of the connection from overflowing the buffers to the receiving host.
- The segments delivered are acknowledged back to the sender upon reception
- Any segments not acknowledged are retransmitted
- Manageable data flow is maintained in order to prevent congestion, overloading and data loss


If too many packets are sent or they are sent too quickly, the computer stores them in the memory section called buffer. If it keeps flooding the device it will eventually start dropping packets and discarding the additional datagramas that arrive.

Network flood control systems actually work pretty well 
The transport can issue a not ready indicator to the sender or source of the flood


A service is considered connection-oriented if:
1. Uses a three-way handshake
2. Uses Sequencing
3. Uses Acknowledgements
4. Uses flow control


ICMP
- Internet Control Message Protocol
- works at the network layer and is used by IP
- Management protocol and messaging service provider for IP
- Messages are carried as packets
- encapsulated within IP datagrams
- provide hosts with information about network programs
- PING uses ICMP echo request and reply messages to check the physical and logical connectivity of machines on an internetwork
- Traceroute uses IP packet Time-To-Live time outs to discover the path a packet takes as it traverses the internetwork
- Block ICMP at firewalls and routers which disables a ping response



ARP
- Address Resolution Protocol
- Finds the hardware address of a host from a known IP address
- When an IP has a datagram to send, it must inform a Network Access Protocol such as ethernet or token ring, of the destinations hardware address on the local network
- It has already been informed by upper-layer protocols of the destination's IP address
- If the Ip doesnt find the destinations host's hardware address in the ARP cache, it uses ARP to find it
- ARP interrogates the local network by sending out a broadcast asking the machine with the specified IP address to reply with its hardware address
- So basically ARP translatesthe software IP address into a hardware address 
- The classic Man-in-the-Middle attack relies on convincing two hosts that the computer in the middle is the other host. This can be accomplished with a domain name spoof if the system is using DNS to identify the other host or address resolution protocol (ARP) spoofing on the LAN.

### Data Encapsulation

When a host transmits data accross a network to another device, the data goes through encapsulation
- Its wrapped with protocol information at each layer of the OSI model
- Each layer communicates only with its peer layer
PDU
- To communicate and exchange information, each layer uses Protocol Data Units (PDU's)
- These hold the control information attached to the data at each layer of the model
- They are usually attached to the header in the front of the data field
- Each PDU attaches to the data by encapsualting it at each layer of the OSI model
- Then its read by the next layer, stripped off and the data is handed to the next level


Works like this...
1. User info is converted to data for transmission on the network
1. Data is converted to degments, and a reliable connection is set up between transmitting and recieving hosts
1. segments are converted to frames for transmission on the local network. 
1. Hardware (ethernet) addresses are used to uniquely identify hosts on the local network
1. Packets or datagrams are converted to bits and a digital encoding and clocking scheme is used.



Packets 

- Packet fields can be modified by attackers for their benefit. 
- The source and destination IP address or ports are very commonly attacked fields in the denial of service and spoofing attacks, as well as in network penetrations.
- Each and every flag and field in a TCP frame and the underlying Ethernet frame can be modified or altered — all for the purpose of network penetration.

- packet crafting and packet spoofing  are very different 
- Spoofing is used by attackers to hide their identities and presence on the network. 
- Spoofing is primarily used to gain network information such as open ports, running services, active hosts, etc., during which time the target host fails to trace the attacker.

- Packet crafting takes a step further by trying to test the presence, functionality or the accuracy of target network’s firewall rules, and intrusion detection systems. 
- Packet crafting requires in-depth knowledge of TCP packets and how they work, and is more of a manually orchestrated attack than a programmatic one. This makes it a technically advanced way of trying to hack into networks.
