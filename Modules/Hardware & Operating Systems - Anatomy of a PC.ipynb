#### 1. Topics/Sub-Topics

    - Three Computing Parts (A+)
        - Hardware
        - Operating System
        - Applications
    - Three Stages of Computing Process (A+)
        1. Input
        2. Processing
        3. Output
    - Nine Steps of Computer/Human Interaction (Troubleshooting Workflow) (A+)
        1. Power up
        2. Processing parts prepare for action
        3. Provide input
        4. Processing parts process commands
        5. Processing parts send output information to output devices
        6. Output devices show output
        7. Repeat steps 3-6
        8. Save your work
        9. Power down

#### 2. Pre-requisites
- A+ Chapter 1-2

#### 3. Desired Outcomes
- Student should be able to identify the main components of a computer
- Student should be able to explain what each component does at a high level
- Student should be able to explain how each component interacts together
- Student should be able to explain where each component goes inside a case
- Student should be able to visually identify the different ports on a PC
- Student should be able to communicate some basic troubleshooting and experiences in working with the components of a computer 
- Student should be able to build a computer/take apart one and put it back together
