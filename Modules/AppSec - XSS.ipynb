### Cross-Site Scripting (XSS)
- More than 30% of sites are still vulernable to this especially if they are old.
- Uses Embeded javascript thats embeded in the html to execute a script.
![Imgur](https://i.imgur.com/oeLbpP4.png)

### Types of XSS
- Stored XSS: Adding something to a database.
- Reflection XSS: A temp response from a server that can trick to run a script.
- DOM based XSS: Could be a query param being rendered to DOM without any filtering.
- Blind XSS(some recognize this): Example, have an internall app that pulls in data from outside source. Can trick this app to evaluting something and take advantage from there.
![Imgur](https://i.imgur.com/oC4d6H7.png)

#### XSS Prevention Quiz
- Plugins should have easy reporting / procedures in place to report issues/vulnerabilities.
- Check to see if they list XSS attacks that they cover and browser requirements.
- Look carefully at permissions when get a new plugin. Should be looking at the code. 
- Always incognito mode. Don't allow extensions incognito mode.
- Browser plugins can XSS if have full accesss. 
- Limit what web applications can do. Ex. limit what can be returned from your database.

  #### Challenge 3: Solution
  - Will be using tool to read the html of the image and replace it with our own.
  - run ```node ./scripts/exif.js read ./examples/squirrel.jpg| pbcopy```. This reads of the image that the browser is rendering.
  - Create a new file scratch.html on root level and paste contents of first command ran into the file. Make sure to keep everything on one line.
  - Run ```node ./scripts/exif.js write ./examples/squirrel.jpg scratch.html```. This will read whatever we put in the scratch.html file and write it to the squirrel.html. Change h1 tag in scratch.html to whatever you want. Will see updated info changed after running command.
  - Change file extension of squirrel.jpg with .html and open in browser. Will see the changes that you made.
  - Shows you can inject code into an image.
