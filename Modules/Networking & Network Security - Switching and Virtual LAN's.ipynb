# Switching and Virtual LANs (VLANs)

- Layer 2 switching is the process of using the hardware addresses of devices on a LAN to segment a network

Layer 2 Switching - How it Works
- Switching breaks up large collision domains into smaller ones and that 
- A collision domain is a network segment with two or more devices sharing the same bandwidth (For you bart!)
    - Another definition: A network segment where data packets may collide with one another while being sent. The collision domain applies particularly in wireless networks, but also affected early versions of Ethernet.
        - This potentially frees us from Ethernet's inherent distance constraints, enabling us to build larger networks
    
- Each port on a switch is actually its own collision domain, you can create a much better ethernet LAN network with switches

- RIP employs processes for preventing network loops from occuring at the network layer which is great, but if you have redundant links between your switches, routing protocols wont do a thing to stop loops from occuring at the data link layer

- Here is why the Spanning Tree Protocol was developed (to put a stop to loops taking place within a layer 2 switched network)

Layer 2 switched networks
- Traditional switching operates at layer 2 of the OSI model
- Packets are sent to a specific switch port based on destination MAC addresses. 
- Routing operates at layer 3, where packets are sent to a specific next-hop IP address, based on destination IP address
- Devices in the same layer 2 segment do not need routing to reach local peers. What is needed however is the destination MAC address which can be resolved through the Address Resolution Protocol (ARP) 

- Switches remove the physical boundry (VLAN's)
- Within a layer 2 switch environment exists a broadcast domain.  
- Any broadcast traffic on a switch will be forwarded out all ports with the exception of the port the broadcast packet arrived on.  
- Broadcasts are contained in the same layer 2 segment, as they do not traverse past a layer 3 boundary.

- Large layer 2 broadcast domains can be susceptible to certain unintended problems, such as broadcast storms, which have the ability to cause network outages.  
- Also, it may be preferable to separate certain clients into different broadcast domains for security and policy reasons.  
- This is when it becomes useful to configure VLANs.  
    - A layer 2 switch can assign VLANs to specific switch ports, which in turn are in different layer 3 subnets, and therefore in different broadcast domains.  
    - VLANs allow for greater flexibility by allowing different layer 3 networks to be sharing the same layer 2 infrastructure.  
    

Switches
- Layer 2 switches and bridges are faster than routers becuase they dont take time to look up the network layer header information, instead they look at the frames hardware address before deciding to forward, flood or drop the frame.
- They create, private, dedicated collision domians and provide independent bandwidth on each port
- Switches provide the following
    - Hardware-based bridging (ASIC)
    - Wire Speed
    - Low latency
    - Low cost
    
- What makes layer 2 switching so efficient is that no modification to the data packet takes place
- The device reads only the frame encapsualting the packet, which makes the switching process conisderably faster
- Also you can create a flatter network designw ith more network segments than you can with traditional routed networks
- Later 2 switchesing increase bandwidth for each user becuase, each connection (interface) into the switch is its own collision domain

Distinct Functions of Layer 2 Switch
1. Address Learning
    - Switch is capable of address learning, which means they rember the souce hardware address of each frame recieved on an interface and enter this information into a MAC database know as a forward/filter table
    - When a switch is initially powered on, the MAC frward/filter table is empty 
    - When a device transmits, and an interface recieves a frame, the switch places the frame's source address in the MAC forward/filter table, which allows it to remeber which interface the sending device is located on
    - The switch then has no choice but to flood the network with this frame out of every port except the source port becuase it has no idea where the destination device is actually located
    - If a device answers this flooded frame and sends a frame back, then the switch wull take the source address from that frame and place a MAC address in its database as well, there associating the newly discovered address with the interface that recieved the frame
    - Now that the switch has noth the relevant MAC addresses in its filtering table, the two devices can make a point-to-point connection
    - The switch doesnt need to flood the frame as it did the first time because now the frames can and will be forwarded only between the two devices recorded in the table

2. Forward/filter decisions
    - when a frame arrives at a switch interface , the destination hardware address is compared to the forward/filter MAC database and the switch makes a forward/filter decision
    - So, if the destination hardware address is known (listed in database), the frame is only sent out to the specified exit interface
    - The switch will not transmit the frame out any interface except the destination interface
        - This preserves bandwidth on the other networks segments and is call frame filtering
    - So again, if the destination MAC address isnt listed in the MAC database, then the frame is flooded out all active interfaces except the interface the frame was recieved on
    - If a device answers the flooded frame, the MAC database is updated with particular interface location

STP
- STP's main task is to stop network loops from occuring on your Layer 2 network (bridges or switches)
- It achieves this by vigilantly monitoring the network to find all links and making sure that no one loops occur by shutting down redundant ones
- Uses the spanning tree algorithm to first create a topology database and then search out and destroy redundant links
- Switches transmit Bridge Protocol Data Units (BPDU's) out all ports so that all links between switches can be found

Port States
- The ports on a switch can transition through 5 different state:
    1. Blocking- Wont forward frames, prevents the use of looped paths, by default
    2. Listening- listens to BPDU's to make sure no loops occur, prepares to forward data frames
    3. Learning- listens and learns all the paths in the switched network
    4. Forwarding- port sends and recieves all data frames on the bridged port
    5. Disabled- administratively disabled and does not participate in forwarding
    
- Ports are usually in either blocking or forwarding
- forwarding port is one that has determined to have the lowest (best) cost to the root bridge
- Once a switch determines the best path to the root bridge, all other redundant ports will be in blocking mode

Convergence
- what happens when all the ports on bridges and switches have transitioned to either forwarding or blocking modes
- during this phase no data will be forwarded until the convergence event is complete
- When STP is convering, all host data stops transmitting
- So if you want to remain on speaking terms with your network's users (or remain employed), you must make sure that your switched network is physically designed really well so that STP can converge quickly and painlessly
- It is important because it ensures that all devices are in either forwarding or blocking mode
- You can make your core switch the STP root
- Because the the typical time to convergence from blocking to forwarding on a switchport is 50 sconds, you can and should disable spanning tree on individual ports

  
    
VLAN's
- Logical grouping of network users and resources connected to administratively defined ports on a switch
- When you create VLAN's you gain the ability to create smaller broadcast domains within layer 2 switched interneworks, by assigning the various ports on the switch to different subnetworks
- A VLAN is treated like its own subnet or broadcast domain, meaning that frames broadcasted onto the network are only switched between the ports logically grouped within the same VLAN
    - So does this mean we dont need router? Maybe Yes and Maybe No- It depends on the specific needs of the network
    - By default, hosts in a specific VLAN cant communicate with hosts that are memebers of another VLAN, so if you want inter-VLAN communication, the answer is yes you still need a router
- Trunking
    - you get to make a single port part of a whole bunch of VLAN's at the same time
    - You can actually set ports up to have a server in two sperate broadcast domains simultaneously so your users wont have corss a Layer 3 device (router) to log in and access it
    - trunk line is a 100 - 1000MBS point-to-point link between 2 switches, switch and router or between switch and server
    - It carries the traffic of multiple VLAN's from 1 -4.094 at a time
- a switch port can belong to only one VLAN 

- Static VLAN's
    - Most common way of configuring, becuase it is the most secure
    - the security stems from the fact that any switch port you've assigned a VLAN association to will always maintain it unless you change the port assignment manually
    - pretty easy to set up and works really well
    
- Security 
    - Security is one of the many capabilities and features that make VLANs so very cool. 
    - If you take the time to build them and create multiple broadcast groups, you get total control over each port and user in return. 
        - So the days when anyone could just plug their workstations into any switch port and gain access to network resources are history. 
    -  Using VLANs gives you control of each port plus whatever resources that port can access. 
        - They can also be configured to inform network management stations of any unauthorized access to network resources
        - You can also place restrictions on hardware addresses, protocols and applications
