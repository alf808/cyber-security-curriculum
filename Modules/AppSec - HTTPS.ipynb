#### HTTPS & Cryptography
- Public key cryptography has both a secret key and public key. The public key is used to create encrypted messages and private key is used to read the messages.
![Imgur](https://i.imgur.com/jGYiRg7.png)
- The server will make its public key + certificate available to anyone that wants to use it just for sending messages that can only be read by the server.
- Cipher suites are the methodology used for that faster symmetrical encryption later on.
- Session keys are one time use  generated by the client. If MITM was happening, couldn't read the data since only server has key to decrypt the clients session key.
![Imgur](https://i.imgur.com/3rfx6Az.png)
- Easy to encrypt one way and impossbile to go the other way. One way encryption
![Imgur](https://i.imgur.com/wtQDnUH.png)

#### TLS Handshake
- Client verifies server certificate using trusted roots that are built in OS or browser. Example, apple root certificate, komoto root cert. So if you use one of the certs, you trust any domain that is signed by these certs.
- Public and private keys you generate are just for the key exchange. Everything in between is done with symmetric encryption cause much faster
![Imgur](https://i.imgur.com/zb5B91o.png)
