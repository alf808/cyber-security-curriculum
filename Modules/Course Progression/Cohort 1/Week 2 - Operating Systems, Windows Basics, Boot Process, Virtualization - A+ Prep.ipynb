     - Difference between BIOS and UEFI
         - BIOS
             - MBR shows location of boot sector
             - Boot sector contains code that points to the file location of the boot manager(bootmgr)
             - Boot manager knows how to launch the OS
         - UEFI
             - Loads the boot manager directly
     - Windows: Unable to locate bootmgr is a standard Windows error message stating which can be signs of corruption or sometimes a virus
