 Security+ Coverage

### Purpose
The Security+ module is a high-level lecture, lab and test prep approach to providing a common vernacular between all current students regardless of background that will set the foundation for the rest of the more technical components of the course and allow students to start preparing for completion and aquisition of the Security+ certification prior to completion of the course

### Desired Outcomes
The desired outcome from the Security+ module are:
    * Students have a common vocabulary and understanding of the various base components and terminology that make up the cybersecurity landscape
    * Students understand and have the resources to continue to self study in preparation for taking the Security+ certification at their leisure
    * Students pass the Security+ certification prior to the completion of the course
    
Students have knowledge of the following main areas of the Security+ SYS-501 test including:

    * Threats, Attacks and Vulnerabilities (21%)
    * Technologies and Tools (22%)
    * Architecture and Design (15%)
    * Identity and Access Management (16%)
    * Risk Management (14%)
    * Cryptography and PKI (12%)
