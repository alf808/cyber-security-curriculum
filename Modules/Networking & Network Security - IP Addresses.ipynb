What is IPv4
- 4th version of the internet protocol but the first deployed in the ARPANET in 1983
- Routes most traffic in the internet today
- Connectionless protocol for use on packet switched networks
- Best effort delivery- does not guarantee delivery, nor does it assure proper sequencing or avoidance of duplicate delivery. 
- These aspects, including data integrity, are addressed by an upper layer transport protocol, such as the Transmission Control Protocol (TCP)
- Functions on the internet layer in TCP/Model and the Network Layer in the OSI model
- TCP functions a layer higher on the Transport Layer

Broadcast Addresses
- 4 different types of broadcast addresses
- Routers by default dont forward broadcasts
    - Layer 2 broadcasts - sent to all hosts on a LAN
        - Dont go past the router
        - hardware address = 6 bytes or 48 bits 0c.43.a4.f3.12.c2
        - Broadcast address would be all bits on in hex FF.FF.FF.FF.FF.FF
    - Broadcast - Sent to all hosts on a network
        - EX: network address of 172.16.0.0 255.255.0.0 would have a broadcast address of 172.16.255.255
        - an example of a broadcast message is an ARP request
        - When a host has a pacet, it knows the logical address (IP) of the destination. to get that packet to its destination, the host needs to forward the packet to a deafult gatewway if the destination resides on a different IP Network. If a destination is on the local networkk the source wil forward the packet directly to the destination. Becuase the source doesnt have the MAC address to which it needs to forward the frame, it sends out a broadcast, which sends out a request for the owner of IP address 192.168.2.3 to please forward MAC address to me.
    - Unicast - sent to a single destination host
        - directed towards a specific host
        - The process of sending a packet from one host to an individual host as shown in Figure 1.
        - For the most part all communication between devices is unicast unless otherwise noted.
        - DHCP request
        - Regardless of whether the destination specified a packet as a unicast, broadcast or multicast; the source address of any packet is always the unicast address of the originating host
        - Used for normal host-to-host communication in both a client/server and a peer-to-peer network. Unicast packets use the address of the destination device as the destination address and can be routed through an internetwork.
        - ex: Your host on a LAN sends out an FF.FF.FF.FF.FF.FF layer 2 broadcast and 255.255.255.255 layer 3 destination broadcast looking for a DHCP serveron the LAN. The router will see that this is a broadcast meant for the DHCP server because it has a destination port number of 67 and will forward the request to the IP address of the DHCP server on another LAN. 
        - So basicallyif your DHCP server IP address is 172.16.10.1 your host just sends out a 255.255.255.255 DHCP client broadcast request, and the router changes that broadcast to the specfic destination address of 172.16.10.1
    - Multicast - Packets sent from a single source and transmitted to many devices on different networks
        - Works by sending messages or data to IP Multicast group addresses
        - routers than forward copies of the packet out every interface that has hosts subscribed to the group address
    
- Classes
    - Class A - For the small number of networks possessing a large number of hosts
        - example IP address 49.22.102.70
        - Network (8 bits) Host.Host.Host
        - First bit is reserved and the 7 other can be manipulated 
        - creators of the internet said that the first bit must always be off or a 0 
        - Maximum number of Class A networks that can be created is 128 (2^7)
        - 00000000 - 01111111 = 127
        - 0 reserved for router and 127 reserved for diagnostics (1-26)
        - 2^24 unique combinations = 16,777,216
        - 10.0.0.0 = all host bits turned off (network address)
        - 10.255.255.255 all host bits turned on is broadcast address
    - Class B - In between A and C
        - example ip address 172.16.30.56
        - Network (8 bits). Network (8 Bits).Host.Host
        - host has 2^16 unique combinations for hosts = 65,534
        - Internet designers said should start with a 1 then a 0
        - 16-2=14 positions to manipulate = 16,384 (2^14) 
        - 10000000 = 128 -  10111111 = 191 (Range for class B)
        - all host bits turned off = 172.16.0.0 (network address)
        - all host bits turned on is the broadcast address 172.16.255.255 (broadcast address)
        - Valid hosts would be network address 172.16.0.1 - 172.16.255.254
    - Class C - Numerous networks with few hosts
        - Network (8 bits). Network (8 bits). Network (8 bits). Host
        - example add (192.168.100.102)
        - first 2 bits in the first octets always turned on but third can never be on
        - First 3 bit postions are 110
        - 3 bytes or 24 bits - 3 reserved positions = 21 positions
        - 2,097,152 class c networks
        - 11000000 = 192 - 11011111 = 223
        - So if you see an address with a range from 192- 223 its a class C 
        - host get 2^8 or 256 (minus all 0's and 1's for a total 254)
        
    - Router Knows that a Class A network always starts with a 0 so it can speed a packet on its way after reading only first bit
    - general rule is when your setting up a corporate network you should use class A
        - EX: if you use 10.0.0.0 /24 you have 65,536 networks each with 254 hosts 
        - EX: Home network you get a 192.168.0.0 with 254 hosts

## Subnetting

- Take 1 network and split it into 6
- Some networks as we know have the potential to have lots of networks

Why do we Subnet
- Reduces Network traffic
     - Routers create broadcast domains so the more broadcast domains the less traffic network traffic on each segment
     - Because routers separate broadcast domains, subdividing networks can improve network performance by eliminating excessive broadcast traffic.
- Simplified Management with smaller connected networks
- A single large WAN that spans a large distance can be very slow and probalamtic
    - Its easier to connect smaller networks together
    
Creating a subnet
- You take bits from the host portion of the IP address and reserve them to define the subnet address
- fewer bits for the hosts so the the more subnets the fewer bits left available for hosts

Subnet Mask
- For the subnet address scheme to work, every machine on the network must know which part of the host address will be used as the subnet address.
    - A subnet mask is assigned to each machine
- A 32 bit value that allows the recipient of IP packets to distinguish the network ID portion of the IP address from the host portion of the IP address


- 192.168.10.       | 10
- Network           |  Host
- Must be identical |  Must be unique
- A portion of the bits identify the network and a portion identify the host
- The Subnet mask is used to determine which portion of the 32-bits identifies the network and which identifies the host
- The network address represents all the devices on the same network.
